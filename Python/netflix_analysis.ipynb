```python
country_counts = (
    df["country"]
    .dropna()
    .str.split(", ")
    .explode()
    .value_counts()
    .head(10)
)

plt.figure(figsize=(8,5))
country_counts.plot(kind="bar")

plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("../Images/top10_countries.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

df["rating"].value_counts().plot(kind="bar")

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.tight_layout()
plt.savefig("../Images/rating_distribution.png")
plt.show()
```

```python
genre_counts = (
    df["listed_in"]
    .str.split(", ")
    .explode()
    .value_counts()
    .head(10)
)

plt.figure(figsize=(8,5))
genre_counts.plot(kind="bar")

plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("../Images/top10_genres.png")
plt.show()
```

```python
year_counts = df["release_year"].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.plot(year_counts.index, year_counts.values)

plt.title("Year-wise Netflix Releases")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.tight_layout()
plt.savefig("../Images/yearwise_releases.png")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df["director"] = df["director"].fillna("Unknown")
```

```python
df["cast"] = df["cast"].fillna("Not Available")
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df = df.dropna(subset=["date_added"])
```

```python
df.isnull().sum()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
df["date_added"] = df["date_added"].str.strip()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')

plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')

plt.title("Top 10 Countries by Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['release_year'].value_counts().sort_index().plot()

plt.title("Content Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    data=df,
    y='rating',
    order=df['rating'].value_counts().index
)

plt.title("Ratings Distribution")

plt.show()
```

```python
genre = df['listed_in'].str.split(',').str[0]

top_genres = genre.value_counts().head(10)

plt.figure(figsize=(10,6))

top_genres.plot(kind='bar')

plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")

plt.show()
```
